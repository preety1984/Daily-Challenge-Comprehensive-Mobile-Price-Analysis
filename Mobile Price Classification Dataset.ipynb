```python
import numpy as np
from scipy import stats

# Select only numeric columns
numeric_data = data.select_dtypes(include=[np.number])

# Central tendency measures
mean = numeric_data.mean()
median = numeric_data.median()
mode = numeric_data.mode().iloc[0]

# Variability measures
range_ = numeric_data.max() - numeric_data.min()
variance = numeric_data.var()
std_dev = numeric_data.std()

# Distribution shapes
skewness = numeric_data.skew()
kurtosis = numeric_data.kurt()

# Print out calculated statistics
print(f"Mean:\n{mean}\n")
print(f"Median:\n{median}\n")
print(f"Mode:\n{mode}\n")
print(f"Range:\n{range_}\n")
print(f"Variance:\n{variance}\n")
print(f"Standard Deviation:\n{std_dev}\n")
print(f"Skewness:\n{skewness}\n")
print(f"Kurtosis:\n{kurtosis}\n")

```

```python
# Hypothesis testing: Example for price range
# Group data by 'price_range'
grouped = data.groupby('price_range')

# Perform t-tests between different price ranges
for name, group in grouped:
    print(f"Price Range {name} Mean:\n{group.mean()}\n")

# Example: t-test between two groups
group1 = data[data['price_range'] == 0]['ram']
group2 = data[data['price_range'] == 1]['ram']
t_stat, p_value = stats.ttest_ind(group1, group2)
print(f"T-statistic: {t_stat}, P-value: {p_value}")

# Feature-target correlations
correlations = data.corr()['price_range']
print(correlations)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms
data.hist(figsize=(11, 10))
plt.show()
```

```python
# Scatter plots
plt.figure(figsize=(10, 6))
plt.scatter(data['battery_power'], data['price_range'])
plt.xlabel('Battery Power')
plt.ylabel('Price Range')
plt.title('Battery Power vs Price Range')
plt.show()
```

```python
# Box plots
plt.figure(figsize=(12, 6))
sns.boxplot(x='price_range', y='ram', data=data)
plt.xlabel('Price Range')
plt.ylabel('RAM')
plt.title('RAM Distribution by Price Range')
plt.show()
```

```python

# Heatmap for correlation
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.show()
```
